```python
import numpy as np
```

```python
melting_point = 933 #K
heat_of_fusion = 1.08e10 #J/m^3
interfacial_energy = 0.065 #J/m^2
undercooling = -200 #K
```

```python
gibbs_energy = heat_of_fusion*(-200)/933
critical_radius = -2*interfacial_energy/gibbs_energy
print("r* = ", critical_radius, "m")
```

```python
critical_free_energy = 16*np.pi*(interfacial_energy**3)/(3*(gibbs_energy**2))
print("G* =", critical_free_energy, "J")
```

```python
import matplotlib.pyplot as plt

plt.xlabel('r (m)')
plt.ylabel('delta G (J)')

xs = []
ys = []
for r in np.arange(0,1e-10,5e-12):
    y = 4*np.pi*interfacial_energy*(r**2)
    xs.append(r)
    ys.append(y)

xs_2 = []
ys_2 = []
for r in np.arange(0,1e-10,5e-12):
    y_2 = (4/3)*np.pi*gibbs_energy*(r**3)
    xs_2.append(r)
    ys_2.append(y_2)
plt.plot(critical_radius, critical_free_energy, linestyle='-', marker='*')
plt.plot(xs, ys, linestyle='-', marker='.')
plt.plot(xs_2, ys_2, linestyle='-', marker='.')

xs_3 = []
ys_3 = []
for r in np.arange(0,1e-10,5e-12):
    y_3 = 4*np.pi*interfacial_energy*(r**2) + (4/3)*np.pi*gibbs_energy*(r**3)
    xs_3.append(r)
    ys_3.append(y_3)
plt.plot(xs_3, ys_3, linestyle='-', marker='_')
```
