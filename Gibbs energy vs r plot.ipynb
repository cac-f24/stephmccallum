```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
melting_point = 933 #K
heat_of_fusion = 1.08e10 #J/m^3
interfacial_energy = 0.065 #J/m^2
undercooling = -200 #K
```

```python
gibbs_energy = heat_of_fusion*(-200)/933
critical_radius = -2*interfacial_energy/gibbs_energy
print("r* = ", critical_radius, "m")
```

```python
critical_free_energy = 16*np.pi*(interfacial_energy**3)/(3*(gibbs_energy**2))
print("G* =", critical_free_energy, "J")
```

```python
plt.xlabel('r (m)')
plt.ylabel('delta G (J)')

xs,xs_2,xs_3 = [],[],[]
ys,ys_2,ys_3 = [],[],[]
for r in np.arange(0,1e-10,5e-12):
    y = 4*np.pi*interfacial_energy*(r**2)
    xs.append(r)
    ys.append(y)
    y_2 = (4/3)*np.pi*gibbs_energy*(r**3)
    xs_2.append(r)
    ys_2.append(y_2)
    y_3 = 4*np.pi*interfacial_energy*(r**2) + (4/3)*np.pi*gibbs_energy*(r**3)
    xs_3.append(r)
    ys_3.append(y_3)

plt.plot(critical_radius, critical_free_energy, linestyle='-', marker='*')
plt.plot(xs, ys, linestyle='-', marker='.')
plt.plot(xs_2, ys_2, linestyle='-', marker='.')
plt.plot(xs_3, ys_3, linestyle='-', marker='_')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from unyt import GJ,mJ,m

melting_point = 933 #K
heat_of_fusion = 10.8*GJ/m**3 #J/m^3
interfacial_energy = 65*mJ/m**2 #J/m^2
undercooling = -200 #K
gibbs_energy = (heat_of_fusion)*(undercooling)/(melting_point)
critical_radius = -2*interfacial_energy/gibbs_energy
print("r* = ", critical_radius)
critical_free_energy = 16*np.pi*(interfacial_energy**3)/(3*(gibbs_energy**2))
print("G* =", critical_free_energy.to('J'))

plt.xlabel('r (m)')
plt.ylabel('delta G (J)')

xs,xs_2,xs_3 = [],[],[]
ys,ys_2,ys_3 = [],[],[]
for r in np.arange(0,1e-10,5e-12):
    y = 4*np.pi*interfacial_energy*((r*m)**2)
    xs.append(r)
    ys.append(y.to('J'))
    y_2 = (4/3)*np.pi*gibbs_energy*((r*m)**3)
    xs_2.append(r)
    ys_2.append(y_2.to('J'))
    y_3 = 4*np.pi*interfacial_energy*((r*m)**2) + (4/3)*np.pi*gibbs_energy*((r*m)**3)
    xs_3.append(r)
    ys_3.append(y_3.to('J'))

plt.plot(critical_radius.to('m'), critical_free_energy.to('J'), linestyle='-', marker='*')
plt.plot(xs, ys,label='4\u03C0\u03B3(r^2)', linestyle='-', marker='.')
plt.plot(xs_2, ys_2,label='(4/3)\u03C0G(r^3)', linestyle='-', marker='.')
plt.plot(xs_3, ys_3, linestyle='-', marker='_')
plt.legend()
```
